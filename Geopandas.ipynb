```python
import geopandas as gpd
from shapely.ops import nearest_points
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
police = gpd.read_file("JK_police.geojson")
school = gpd.read_file("JK_school.geojson")
hospital = gpd.read_file("JK_hospital.geojson")
```

```python
display(school)
display(police)
display(hospital)
```

```python
# List of important columns
featured_columns = ['name', 'id', 'geometry']
school_imp = school[featured_columns]
hospital_imp = hospital[featured_columns]
police_imp = police[featured_columns]
```

```python
school_imp
```

```python
# Create a GeoDataFrame directly
school_gdf = gpd.GeoDataFrame(school, geometry='geometry')
hospital_gdf = gpd.GeoDataFrame(hospital, geometry='geometry')
police_gdf = gpd.GeoDataFrame(police, geometry='geometry')
```

```python
police_gdf
```

```python
# Define a suitable CRS for area calculation (e.g., EPSG:3857 for Web Mercator projection)
#CRS - Coordinate Reference System
projected_crs = 'EPSG:3857'  # You can choose an appropriate CRS

# Reproject the GeoDataFrame to the projected CRS
school_gdf = school_gdf.to_crs(projected_crs)
hospital_gdf = hospital_gdf.to_crs(projected_crs)
police_gdf = police_gdf.to_crs(projected_crs)
```

```python
# Calculate the average size of parks in the projected CRS
school_gdf['school_area'] = school_gdf['geometry'].area
hospital_gdf['hospital_area'] = hospital_gdf['geometry'].area
police_gdf['police_area'] = police_gdf['geometry'].area
```

```python
school_gdf.head()
```

```python
hospital_gdf.head()
```

```python
# Empty lists to store the results
nearest_distances = []
nearest_school_names = []

for neighborhood in hospital_gdf.iterrows():
    distances = []
    school_names = []
    for school in school_gdf.iterrows():
        nearest = nearest_points(neighborhood[1].geometry, school[1].geometry)
        distances.append(nearest[0].distance(nearest[1]))
        school_names.append(school[1]['name'])  # Assuming the park name is in the 'name' column of the park_gdf
    nearest_distances.append(np.mean(distances))
    nearest_school_names.append(school_names[np.argmin(distances)])

# Update the DataFrame with the nearest distances and park names
hospital_gdf['Avg Distance to Nearest School'] = nearest_distances
hospital_gdf['Nearest School Name'] = nearest_school_names
```

```python
# Calculate the average size of schools
school_gdf['school_area'] = school_gdf['geometry'].area
average_school_size = school_gdf['school_area'].mean()

print("Total number of schools:", len(school_gdf))
print("Average size of schools (in square meters):", average_school_size)
```

```python
# Calculate the average size of hospitals
hospital_gdf['hospital_area'] = hospital_gdf['geometry'].area
average_hospital_size = hospital_gdf['hospital_area'].mean()

print("Total number of hospitals:", len(hospital_gdf))
print("Average size of hospitals (in square meters):", average_hospital_size)
```

```python
# Calculate the average size of police stations
police_gdf['police_area'] = police_gdf['geometry'].area
average_police_size = police_gdf['police_area'].mean()

print("Total number of Police station:", len(police_gdf))
print("Average size of police station (in square meters):", average_police_size)
```

```python
# Perform a spatial join
joined = gpd.sjoin(police_gdf, school_gdf, how='left', op='intersects')

# Check the columns in the joined dataframe
print(joined.columns)

# Determine the correct columns to use for school area per neighborhood
school_area_column = 'school_area_left' if 'school_area_left' in joined.columns else 'index_right'

# Calculate the number of schools and total school area per neighborhood
school_per_neighborhood = joined.groupby('name_left').size()

# Create a DataFrame with the calculated metrics
neighborhood_metrics = pd.DataFrame({
    'Neighborhood': school_per_neighborhood.index,
    'Number of schools': school_per_neighborhood.values,

})
neighborhood_metrics.head()
```

```python
fig = px.bar(neighborhood_metrics, x='Number of schools', y='Neighborhood', color='Neighborhood',
             labels={'Number of Schools': 'Number of Schools'}, title='Number of Schools in Each Neighborhood Police station')
fig.update_layout(xaxis_tickangle=-45)  # Rotate x-axis labels for better readability
fig.show()
```
